```python
import time
```

```python
def suma_rec(n):
    if n>0:
        return int(n+suma_rec(n-1))
    else:
        return 0
```

```python
suma_rec(6)
```

```python
for i in range(0,100):
    inicio=time.time()

    suma_rec(100)

    final = time.time()

    ejecucion = []

    ejecucion.append(final-inicio)

    print(f"Tiempo: {ejecucion}")

suma=0
for t in ejecucion:
    suma = suma + t
    
promedio=suma/100
print(f"tiempo promedio de ejecucion: {promedio}")
```

```python
def suma_formula(n):
    resultado = (n*(n+1))/2
    return resultado
```

```python
suma_formula(100)
```

```python
for i in range(0,100):
    inicio=time.time()

    suma_formula(100)

    final = time.time()

    ejecucion = []

    ejecucion.append(final-inicio)

    print(f"Tiempo: {ejecucion}")

suma=0
for t in ejecucion:
    suma = suma + t
    
promedio=suma/100
print(f"tiempo promedio de ejecucion: {promedio}")
```
